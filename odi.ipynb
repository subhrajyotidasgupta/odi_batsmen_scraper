```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm
```

```python
#The 'Invisible Cloak' to not face any consequences from server
headers = {'user-agent':'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.77 Safari/537.36'}
```

```python
player_names = []
years = []
runs = []

#scrapper function
def scrapper(page):
    soup = BeautifulSoup(page.text, 'html.parser')
    results = soup.find('tbody')
    #getting player_names
    links = results.findAll('a')
    del links[1::2]
    for link in links:
        player_names.append(link.getText())
    #getting player's span of years played
    years_noisy = results.findAll('td', attrs={'class': 'left'})
    del years_noisy[0:len(years_noisy):2]
    for y in years_noisy:
        years.append(y.getText())
    #getting player runs
    runs_noisy = results.findAll('b')
    for r in runs_noisy:
        runs.append(r.getText())
```

```python
root_url = "http://stats.espncricinfo.com/ci/engine/stats/index.html?class=2;template=results;type=batting"

#Progress Bar
pbar = tqdm(range(1, 52))

for i in pbar:
    #scrapping the first(root) page
    if i==1:
        page = requests.get(root_url, headers=headers)
        scrapper(page)
    #scrapping the consecutive pages
    else:
        rest_urls = root_url + ';page=' + str(i)
        page = requests.get(rest_urls, headers=headers)
        scrapper(page)
    pbar.set_description(f'Page [{i}] scraped!')
```

```python
#creating the dataframe from the data
df = pd.DataFrame()
df['Players'] = player_names
df['Years'] = years
df['Runs'] = runs

df
```
